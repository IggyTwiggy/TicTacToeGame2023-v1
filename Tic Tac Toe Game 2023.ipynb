```python
# CELL SELECTION

def player_choice(board):
    
    selection = False
        
    while selection == False:
        selection = input("Please select your next position on the board (1-9):  ")
    
        if selection.isdigit():
            if int(selection) in range (1, 10):
                selection_int = int(selection)
                
                if space_check(board, selection_int) == True:
                    return selection_int
                else:
                    clear_output()
                    display_board(board)                    
                    print("\nYou have selected a position that is filled. Try again, friend :)")
                    selection = False
                
            else:
                clear_output()
                display_board(board) 
                print("\nInvalid selection. Please choose a number from 1 to 9")
                selection = False
        else:
            clear_output()
            display_board(board)
            print("\nInvalid selection. Please choose a number from 1 to 9")
            selection = False
                        
        
```

```python
# IS CELL AVAILABLE?

def space_check(board, position):
    
    if board[position] == " ":
        return True
    else:
        #print("Sorry, this position is already taken")
        return False
    
```
